# **CMU Movie Summary Corpus**


Le **CMU Movie Summary Corpus** est une collection complète de résumés de films et de métadonnées associées, utilisée principalement pour des travaux de recherche en traitement du langage naturel et en analyse de données cinématographiques. Voici un aperçu détaillé des composants du dataset :

### **Résumé des Données**

1. **plot_summaries.txt.gz [29 M]**
   
   - **Contenu** : Résumés de l'intrigue de 42 306 films.
   - **Source** : Extraits du dump de Wikipédia en langue anglaise daté du 2 novembre 2012.
   - **Format** : Chaque ligne commence par l'ID du film sur Wikipédia, suivi du résumé de l'intrigue.

2. **corenlp_plot_summaries.tar.gz [628 M, téléchargement séparé]**
   
   - **Contenu** : Les mêmes résumés d'intrigue traités par le pipeline Stanford CoreNLP.
   - **Fonctionnalités** : Étiquetage grammatical, analyse syntaxique, reconnaissance d'entités nommées (NER) et résolution des coréférences.
   - **Format des fichiers** : Chaque fichier est nommé avec l'ID du film correspondant sur Wikipédia.

### **Métadonnées**

3. **movie.metadata.tsv.gz [3.4 M]**
   
   - **Contenu** : Métadonnées pour 81 741 films.
   - **Source** : Extrait du dump Freebase du 4 novembre 2012.
   - **Format** : Fichier tabulé avec les colonnes suivantes :
     1. ID du film sur Wikipédia
     2. ID du film sur Freebase
     3. Nom du film
     4. Date de sortie
     5. Revenus au box-office
     6. Durée du film
     7. Langues du film (paires ID Freebase: nom)
     8. Pays de production (paires ID Freebase: nom)
     9. Genres du film (paires ID Freebase: nom)

4. **character.metadata.tsv.gz [14 M]**
   
   - **Contenu** : Métadonnées pour 450 669 personnages alignés avec les films mentionnés précédemment.
   - **Source** : Extrait du dump Freebase du 4 novembre 2012.
   - **Format** : Fichier tabulé avec les colonnes suivantes :
     1. ID du film sur Wikipédia
     2. ID du film sur Freebase
     3. Date de sortie du film
     4. Nom du personnage
     5. Date de naissance de l'acteur
     6. Sexe de l'acteur
     7. Taille de l'acteur (en mètres)
     8. Ethnicité de l'acteur (ID Freebase)
     9. Nom de l'acteur
     10. Âge de l'acteur au moment de la sortie du film
     11. ID de la correspondance personnage/acteur sur Freebase
     12. ID du personnage sur Freebase
     13. ID de l'acteur sur Freebase

### **Données de Test**

- **tvtropes.clusters.txt**
  
  - **Contenu** : 72 types de personnages issus de tvtropes.com, avec 501 instances de ces types.
  - **Référence** : Le champ ID correspond à l'ID de correspondance personnage/acteur sur Freebase dans `character.metadata.tsv`.

- **name.clusters.txt**
  
  - **Contenu** : 970 noms de personnages uniques utilisés dans au moins deux films différents, avec 2 666 instances de ces types.
  - **Référence** : Le champ ID correspond à l'ID de correspondance personnage/acteur sur Freebase dans `character.metadata.tsv`.

## OMDB (https://www.omdbapi.com/)

De manière générale, les informations fournies sur un film incluent :

1. **Titre et Année de Sortie** : Le nom du film et l'année où il est sorti.
2. **Classification** : Indique si le film est adapté pour certains publics (par exemple, "R" pour Restricted).
3. **Date de Sortie en Salle** : Date précise de sortie.
4. **Durée** : Longueur du film.
5. **Genres** : Types de film (ex. action, horreur, science-fiction).
6. **Réalisateur et Scénariste(s)** : Les personnes responsables de la direction et de l'écriture.
7. **Acteurs Principaux** : Noms des acteurs principaux.
8. **Résumé (Plot)** : Brève description de l'intrigue.
9. **Langue et Pays** : Langue principale et pays d'origine.
10. **Récompenses** : Nominations ou prix reçus.
11. **Affiche (Poster)** : Lien vers l'image de l'affiche.
12. **Évaluations** : Notes sur différents sites (IMDb, Rotten Tomatoes, etc.).
13. **Scores Spécifiques** : Par exemple, Metascore ou note IMDb.
14. **Nombre de Votes** : Indique la popularité basée sur les avis.
15. **Box Office** : Revenus générés par le film.
16. **Production et Site Web** : Informations sur la société de production et un lien vers le site officiel, si disponibles.

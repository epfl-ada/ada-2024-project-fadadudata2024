# **CMU Movie Summary Corpus**


Le **CMU Movie Summary Corpus** est une collection complète de résumés de films et de métadonnées associées, utilisée principalement pour des travaux de recherche en traitement du langage naturel et en analyse de données cinématographiques. Voici un aperçu détaillé des composants du dataset :

### **Résumé des Données**

1. **plot_summaries.txt.gz [29 M]**
   
   - **Contenu** : Résumés de l'intrigue de 42 306 films.
   - **Source** : Extraits du dump de Wikipédia en langue anglaise daté du 2 novembre 2012.
   - **Format** : Chaque ligne commence par l'ID du film sur Wikipédia, suivi du résumé de l'intrigue.

2. **corenlp_plot_summaries.tar.gz [628 M, téléchargement séparé]**
   
   - **Contenu** : Les mêmes résumés d'intrigue traités par le pipeline Stanford CoreNLP.
   - **Fonctionnalités** : Étiquetage grammatical, analyse syntaxique, reconnaissance d'entités nommées (NER) et résolution des coréférences.
   - **Format des fichiers** : Chaque fichier est nommé avec l'ID du film correspondant sur Wikipédia.

### **Métadonnées**

3. **movie.metadata.tsv.gz [3.4 M]**
   
   - **Contenu** : Métadonnées pour 81 741 films.
   - **Source** : Extrait du dump Freebase du 4 novembre 2012.
   - **Format** : Fichier tabulé avec les colonnes suivantes :
     1. ID du film sur Wikipédia
     2. ID du film sur Freebase
     3. Nom du film
     4. Date de sortie
     5. Revenus au box-office
     6. Durée du film
     7. Langues du film (paires ID Freebase: nom)
     8. Pays de production (paires ID Freebase: nom)
     9. Genres du film (paires ID Freebase: nom)

4. **character.metadata.tsv.gz [14 M]**
   
   - **Contenu** : Métadonnées pour 450 669 personnages alignés avec les films mentionnés précédemment.
   - **Source** : Extrait du dump Freebase du 4 novembre 2012.
   - **Format** : Fichier tabulé avec les colonnes suivantes :
     1. ID du film sur Wikipédia
     2. ID du film sur Freebase
     3. Date de sortie du film
     4. Nom du personnage
     5. Date de naissance de l'acteur
     6. Sexe de l'acteur
     7. Taille de l'acteur (en mètres)
     8. Ethnicité de l'acteur (ID Freebase)
     9. Nom de l'acteur
     10. Âge de l'acteur au moment de la sortie du film
     11. ID de la correspondance personnage/acteur sur Freebase
     12. ID du personnage sur Freebase
     13. ID de l'acteur sur Freebase

### **Données de Test**

- **tvtropes.clusters.txt**
  
  - **Contenu** : 72 types de personnages issus de tvtropes.com, avec 501 instances de ces types.
  - **Référence** : Le champ ID correspond à l'ID de correspondance personnage/acteur sur Freebase dans `character.metadata.tsv`.

- **name.clusters.txt**
  
  - **Contenu** : 970 noms de personnages uniques utilisés dans au moins deux films différents, avec 2 666 instances de ces types.
  - **Référence** : Le champ ID correspond à l'ID de correspondance personnage/acteur sur Freebase dans `character.metadata.tsv`.

## OMDB (https://www.omdbapi.com/)

De manière générale, les informations fournies sur un film incluent :

1. **Titre et Année de Sortie** : Le nom du film et l'année où il est sorti.
2. **Classification** : Indique si le film est adapté pour certains publics (par exemple, "R" pour Restricted).
3. **Date de Sortie en Salle** : Date précise de sortie.
4. **Durée** : Longueur du film.
5. **Genres** : Types de film (ex. action, horreur, science-fiction).
6. **Réalisateur et Scénariste(s)** : Les personnes responsables de la direction et de l'écriture.
7. **Acteurs Principaux** : Noms des acteurs principaux.
8. **Résumé (Plot)** : Brève description de l'intrigue.
9. **Langue et Pays** : Langue principale et pays d'origine.
10. **Récompenses** : Nominations ou prix reçus.
11. **Affiche (Poster)** : Lien vers l'image de l'affiche.
12. **Évaluations** : Notes sur différents sites (IMDb, Rotten Tomatoes, etc.).
13. **Scores Spécifiques** : Par exemple, Metascore ou note IMDb.
14. **Nombre de Votes** : Indique la popularité basée sur les avis.
15. **Box Office** : Revenus générés par le film.
16. **Production et Site Web** : Informations sur la société de production et un lien vers le site officiel, si disponibles.

### Creation of a colomn continent and select only American and European movies

### Awards

```python
import re

# Function to extract the number of awards from the Awards column
def extract_awards(awards_str):
    if pd.isna(awards_str):
        return 0, 0, 0
    oscar = 0
    nomination = 0
    win = 0
    oscar_match = re.search(r'Won (\d+) Oscar', awards_str)
    if oscar_match:
        oscar = int(oscar_match.group(1))
    nomination_match = re.search(r'(\d+) nomination', awards_str)
    if nomination_match:
        nomination = int(nomination_match.group(1))
    win_match = re.search(r'(\d+) win', awards_str)
    if win_match:
        win = int(win_match.group(1))
    return oscar, nomination, win

# Apply the function to the Awards column and create new columns
merged_movie_metadata[['Oscar', 'Nomination_Awards', 'Win_Awards']] = merged_movie_metadata['Awards'].apply(lambda x: pd.Series(extract_awards(x)))

# Display the first few rows to verify the changes
display(merged_movie_metadata[['Awards', 'Oscar', 'Nomination_Awards', 'Win_Awards']].head())


```

```python
import json

# Function to extract ratings from the Ratings column
def extract_ratings(ratings_str):
    if pd.isna(ratings_str) or ratings_str == '[]':
        return {}
    ratings_list = json.loads(ratings_str.replace("'", '"'))
    ratings_dict = {rating['Source']: rating['Value'] for rating in ratings_list}
    return ratings_dict

# Apply the function to the Ratings column
merged_movie_metadata['Ratings_Dict'] = merged_movie_metadata['Ratings'].apply(extract_ratings)

# Extract ratings from different sources
merged_movie_metadata['Internet_Movie_Database_Rating'] = merged_movie_metadata['Ratings_Dict'].apply(lambda x: x.get('Internet Movie Database', None))
merged_movie_metadata['Rotten_Tomatoes_Rating'] = merged_movie_metadata['Ratings_Dict'].apply(lambda x: x.get('Rotten Tomatoes', None))
merged_movie_metadata['Metacritic_Rating'] = merged_movie_metadata['Ratings_Dict'].apply(lambda x: x.get('Metacritic', None))

# Drop the temporary Ratings_Dict column
merged_movie_metadata = merged_movie_metadata.drop(columns=['Ratings_Dict'])

# Display the first few rows to verify the changes
display(merged_movie_metadata[['Movie_Name', 'Internet_Movie_Database_Rating', 'Rotten_Tomatoes_Rating', 'Metacritic_Rating']].head())
```

```python

# Function to convert ratings to a scale of 10
def convert_to_scale_of_10(rating):
    if pd.isna(rating):
        return None
    if isinstance(rating, str):
        if '%' in rating:
            return float(rating.replace('%', '')) / 10
        if '/10' in rating:
            return float(rating.replace('/10', ''))
        if '/100' in rating:
            return float(rating.replace('/100', '')) / 10
    return float(rating)

# Apply the function to the rating columns
merged_movie_metadata['Internet_Movie_Database_Rating'] = merged_movie_metadata['Internet_Movie_Database_Rating'].apply(convert_to_scale_of_10)
merged_movie_metadata['Rotten_Tomatoes_Rating'] = merged_movie_metadata['Rotten_Tomatoes_Rating'].apply(convert_to_scale_of_10)
merged_movie_metadata['Metacritic_Rating'] = merged_movie_metadata['Metacritic_Rating'].apply(convert_to_scale_of_10)

# Display the first few rows to verify the changes
display(merged_movie_metadata[['Movie_Name', 'Internet_Movie_Database_Rating', 'Rotten_Tomatoes_Rating', 'Metacritic_Rating']].head())
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Function to create heatmap for a given continent
def create_heatmap(continent):
    # Filter data for the given continent
    continent_data = exploded_df[exploded_df['Continents'].apply(lambda x: continent in x)]
    
    # Group by comedy genres and rated, then calculate the mean box office revenue
    heatmap_data = continent_data.groupby(['comedy_genres', 'Rated'])['Internet_Movie_Database_Rating'].mean().unstack()
    
    # Plot the heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Internet_Movie_Database_Rating'})
    plt.title(f'Heatmap of Internet_Movie_Database_Rating for {continent}')
    plt.xlabel('Rated')
    plt.ylabel('Comedy Genres')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.show()

# Create heatmaps for each continent
for continent in ['America', 'Europe', 'Both']:
    create_heatmap(continent)
```
